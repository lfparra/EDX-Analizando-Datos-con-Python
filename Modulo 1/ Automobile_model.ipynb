```python
# df[df['price'] =="?"]

df.drop(df[df['price'] =="?"].index, inplace = True)

# Eliminar filas según por una condición
```

```python
df['price'] = df['price'].astype(float)
```

```python
df['normalized-losses'] =="?"
```

```python
df['normalized-losses'].replace('?',np.nan, inplace = True)
```

```python
df['normalized-losses'] = df['normalized-losses'].astype(float)
```

```python
df['normalized-losses'].mean()
```

```python
df['normalized-losses'].replace(np.nan, df['normalized-losses'].mean(), inplace = True)
```

```python
df['city-mpg'] = 235/df['city-mpg'] # Conversión a L/100Km
```

```python
df.rename(columns={'city-mpg': 'city-L/100km'}, inplace=True)
```

```python
df['length_simple_scaling'] = df['length']/df['length'].max() # 
```

```python
bins = np.linspace(min(df['price']), max(df['price']), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['price_bined'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)
df['price_bined']
```

```python
df_fuel_type = pd.get_dummies(df['fuel-type'])
df_fuel_type
```

```python
df['diesel'] = df_fuel_type['diesel']
df['gas'] = df_fuel_type['gas']
```
